### Data

```python
import pandas as pd
import numpy as np

np.random.seed(42)

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = [12000, 15000, 13500, 17000, 16000, 19000,
         18500, 21000, 20000, 22500, 24000, 26000]

subjects = ['Math', 'Science', 'English', 'History', 'Art']
scores   = [78, 85, 72, 90, 65]

study_hours = np.random.uniform(1, 10, 100)
exam_scores = study_hours * 8 + np.random.normal(0, 5, 100)

product_sales = np.random.normal(loc=50, scale=10, size=200)
```

### Task 1 - Line Plot: Monthly Sales Trend

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.plot(months, sales, color='green', marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("\nMonths")
plt.ylabel("Sales\n")
plt.show()
```

### Task 2 - Bar Plot: Subject Score Comparison

```python
plt.figure(figsize=(6,4))
plt.bar(subjects, scores)
plt.title("Subject Score Comparison")
plt.xlabel("\nSubjects")
plt.ylabel("Scores\n")
plt.show()
```

### Task 3 - Scatter Plot: Study Hours vs Exam Score

```python
import seaborn as sns

plt.figure(figsize=(6,4))
sns.scatterplot(x=study_hours, y=exam_scores, hue=exam_scores, alpha=0.7)
plt.title("Study Hours vs Exam Score")
plt.xlabel("\nStudy Hours")
plt.ylabel("Exam Scores\n")
plt.show()
```

### Task 4 - Box Plot: Product Sales Distribution

```python
plt.figure(figsize=(8, 4))
sns.boxplot(x=product_sales)
plt.title("Product Sales Distribution")
plt.xlabel("\nSales")
plt.show()
```
